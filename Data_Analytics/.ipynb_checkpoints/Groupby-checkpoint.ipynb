```python
import pandas as pd

data = pd.read_excel("Data/E_data1.xlsx")
print(data)
```

```python
gp = data.groupby("Department").agg({"Gender":"count"})
print(gp)
```

```python
gp = data.groupby("Department").agg({"E_ID":"count"})
print(gp)
```

```python
gp = data.groupby(["Department","Gender"]).agg({"E_ID":"count"})
print(gp)
```

```python
gp1 = data.groupby(["Country","Gender"]).agg({"Age":"mean"})
print(gp1)
```

```python
gp2 = data.groupby("Country").agg({"Annual_Salary":"mean"})
print(gp2)
```
