```python
import numpy as np
import statistics as stats

baked_food = [200,220,250,230,200,260,240,230]

a = np.array(baked_food)

print(np.mean(baked_food)) #sum of all the values/number of values
# print(np.median(baked_food)) #middle value

#-----------Mode---------------
# print(stats.mode(baked_food)) #most common value

#-------- Standard Deviation ---------------------
print(np.std(baked_food)) #average distance from mean

#*HINT-(if you have Standard Deviation is: 20.27 then you find out variance is: 20.27*20.27=410.87)

#------- Variance------------------------------------
print(np.var(baked_food)) #how spread out the values are
# print(55.68**2) # here: example: stander dervation result = 55.68 if square then find variance look at

#*HINT-(if you have VARIANCE is: 410.93 then you find out Standard Deviation : √410.93 = 20.27))

```

```python
# -1 represent inversely proportional relationship
#  1 represent proportional relationship
#  0 means no relationship
tobacoo_consumption = [30,50,10,30,50,40]
deaths = [100,120,70,100,120,112]

print(np.corrcoef([tobacoo_consumption,deaths]))
```

```python
baked_food = [200,300,250,230,200,280,270,288]
sales =      [20, 35, 25, 23, 19, 30, 28, 32]

print(np.corrcoef([baked_food, sales]))

```

```python
'''| Correlation value | Meaning              |
| ----------------- | -------------------- |
| +1.0              | Perfect positive     |
| +0.9 to +1.0      | Very strong positive |
| +0.7 to +0.9      | Strong positive      |
| +0.3 to +0.7      | Moderate positive    |
| 0.0 to +0.3       | Weak positive        |
| −0.3 to 0.0       | Weak negative        |
| −0.7 to −0.3      | Moderate negative    |
| −0.9 to −0.7      | Strong negative      |
| −1.0              | Perfect negative     |'''
```

```python
#EXAMPLE 1: FOR CO-RELATION

study_hours = [1, 2, 3, 4, 5]
marks =       [40, 50, 60, 70, 80]

import numpy as np

corr = np.corrcoef(study_hours, marks)
print(corr)


'''Interpretation

Correlation = 1

Perfect positive relationship

More study → higher marks (every time)'''
```

```python
import numpy as np

exercise_hours = [1, 2, 3, 4, 5]
weight =         [80, 78, 76, 74, 72]

corr = np.corrcoef(exercise_hours, weight)
print(corr)

'''Interpretation

Correlation = −1

More exercise → less weight

Perfect negative relationship'''

```
