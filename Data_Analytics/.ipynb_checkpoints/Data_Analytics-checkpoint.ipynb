```python
import numpy as np
arr3 = np.array([[10,20,40],[50,60,70],[15,25,45]])
print(np.shape(arr3))  # returns the dimensions of the array (rows, columns)
print(np.size(arr3))   # returns total number of elements in the array
print(np.ndim(arr3))   # returns number of dimensions (1D, 2D, 3D, etc.)
print(arr3.dtype)      # returns the data type of elements in the array
```

```python
import numpy as np
a = ([[1,2,3,9],[4,5,6,0]])
arr = np.array(a)
print(len(arr))  # number of nested values
print(arr.shape) # returns the dimensions of the array (rows, columns)
print(arr.size)  # returns total number of elements in the array
print(arr.ndim)  # returns number of dimensions (1D, 2D, 3D, etc.)
print(type(arr))  # type of variable
print(arr.dtype)     # returns the data type of elements in the array
print(arr.astype(str)) #returns a new array with all elements converted to float
```

```python
# Concatenate
import numpy as np

arr1 = np.array([[10,20],[45,30]])
arr2 = np.array([[40,50],[55,60]])

print(np.concatenate([arr1,arr2]))
```

```python
print(np.hstack([arr1,arr2])) # horizontal concatenation
```
