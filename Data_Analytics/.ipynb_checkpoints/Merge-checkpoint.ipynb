```python
#Example : 1
import pandas as pd


data1 = {"Emp Id":["E01","E02","E03","E04","E05"],
         "Names":["Ram","Shyam","Rahul","Ramu","Ramiya"],
         "Age":[34,34,29,28,32]}


data2 = {"Emp Id":["E01","E02","E03","E04","E05"],
         "Salary":[45000,56000,34000,50000,40000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)
print()
print(df2)
```

```python
print(pd.merge(df1,df2, on = "Emp Id")) #merge opeation
```

```python
#Example:2
import pandas as pd


data1 = {"Emp Id":["E01","E02","E03","E07","E05"],
         "Names":["Ram","Shyam","Rahul","Ramu","Ramiya"],
         "Age":[34,34,29,28,32]}


data2 = {"Emp Id":["E01","E06","E03","E04","E05"],
         "Salary":[45000,56000,34000,50000,40000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)
print()
print(df2)
```

```python
print(pd.merge(df1,df2, on = "Emp Id")) #merge opeation only give you matching values
```

```python
print(pd.merge(df1,df2, on = "Emp Id", how="left")) #merge opeation with left
```

```python
print(pd.merge(df1,df2, on = "Emp Id", how = "right"))  #merge opeation with right
```

```python
#EXAMPLE: 3
import pandas as pd


data1 = {"Emp Id":["E01","E02","E03","E07","E05"],
         "Names":["Ram","Shyam","Rahul","Ramu","Ramiya"],
         "Age":[34,34,29,28,32]}


data2 = {"Emp Id":["E06","E07","E08","E09","E10"],
         "Names":["Guddu","Bantu","Santu","Golu","Sonu"],
         "Age":[34,34,29,28,32]}


df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(pd.concat([df1,df2]))
```
