```python
import matplotlib.pyplot as plt

y = [98,67,88,95,88]
x = ["Part1","Part2","Part3","Part4","Part5"]

colors = ["#FF5733", "#33FF57", "#3357FF", "#F333FF", "#33FFF5"]
plt.bar(x,y, color = colors, edgecolor = "black")
plt.title("Harry Potter Parts popularity", fontsize = 13)
plt.xlabel("Parts of Harry Potter",fontsize = 12.5)
plt.ylabel("Popularity of Harry Potter",fontsize = 12.5)
plt.show()
```

```python
#Here use groupby to aggregate the amounts based on payment modes
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("Data/expense.xlsx")
# print(data)
#Here look the actual amounts and payment modes in the excel file to understand the bar plot better
df  = pd.DataFrame(data)
groupby = df.groupby("Payment_Mode")["Amount"].sum()
print(groupby)
plt.bar(groupby.index, groupby.values, color = ["#FF5733", "#33FF57", "#3357FF"])
plt.title("Total Expense by Payment Mode", fontsize = 13)
plt.xlabel("Payment Mode", fontsize = 12.5)
plt.ylabel("Total Amount Spent", fontsize = 12.5)
plt.show()
```
