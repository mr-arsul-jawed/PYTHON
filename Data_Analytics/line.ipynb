```python
import matplotlib.pyplot as plt

x = ["Day1", "Day2", "Day3", "Day4", "Day5"]
y = [300, 450, 200, 600, 800]
y1 = [280, 400, 250, 550, 700]

plt.plot(x, y, marker='o', linestyle='--', color='r', label='Week 1 ',alpha=0.7)
plt.plot(x, y1, marker='s', linestyle='-', color='b', label='week 2')
plt.title("Daily Expenses Comparison")
plt.xlabel("Days")
plt.ylabel("Amount")
plt.legend()
plt.show()
```

```python
#Here use directly the excel file to plot the bar plot
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("Data/expense.xlsx")
# print(data)

df  = pd.DataFrame(data)
# print(df)


plt.plot(df["Date"], df["Amount"], marker='o', linestyle='-', color='g')
plt.title("Daily Expenses Over Time", fontsize = 14)
plt.xlabel("Date", fontsize = 12)
plt.ylabel("Amount Spent", fontsize = 12)
plt.show()

```

```python
#Here use groupby to aggregate the amounts based on payment modes
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("Data/expense.xlsx")
# print(data)

df  = pd.DataFrame(data)
# print(df)
    
groupby = df.groupby("Date")["Amount"].sum()
# print(groupby)
plt.plot(groupby.index, groupby.values, marker='o', linestyle='-', color='r')
plt.title("Total Expense by Date", fontsize = 13)
plt.xlabel("Date", fontsize = 12.5)
plt.ylabel("Total Amount Spent", fontsize = 12.5)
plt.show()
```

```python
#Example 2: Here use groupby to aggregate the amounts based on payment modes
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel("Data/expense.xlsx")

df = pd.DataFrame(data)
# print(df)

groupby = df.groupby("Category")["Amount"].sum()
print(groupby)
plt.plot(groupby.index, groupby.values, marker='o', linestyle='--', color='r')
plt.title("Total Expense by Category", fontsize = 12)
plt.xlabel("Category", fontsize = 11)
plt.ylabel("Total Amount Spent", fontsize = 11)
plt.show()
```
