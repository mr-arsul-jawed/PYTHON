```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,10,50)
y = np.random.randint(1,100,50)
color = np.random.randint(10,100,50)
size = np.random.randint(10,100,50)

plt.scatter(x, y, c= color, s=size, marker='*', cmap='viridis')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_excel("Data/expense.xlsx")
# print(data)

df  = pd.DataFrame(data)
# print(df)



plt.scatter(df["Date"], df["Amount"], marker = '*', color = "r")
plt.colorbar()
plt.title("Expense Scatter Plot", fontsize = 13)
plt.xlabel("Date", fontsize = 12.5)
plt.ylabel("Amount Spent", fontsize = 12.5)
plt.show()
```
